# 2D Projection Technique

**Description**: Projecting rows of a dataset from a usually high-dimensional original space into a more
visually understandable lower-dimensional space, ideally two or three dimensions. Techniques to
achieve this include Principal Components Analysis (PCA), Multidimensional Scaling (MDS), t-
Distributed Stochastic Neighbor Embedding (t-SNE), and Autoencoder Networks.

**Suggested usage**: The key idea is to represent the rows of a dataset in a meaningful low-dimensional
space. Datasets containing images, text, or even business data with many variables can be difficult to
visualize as a whole. These projection techniques enable high-dimensional datasets to be projected
into representative low-dimensional spaces and visualized using the trusty old scatter plot technique. A
high-quality projection visualized in a scatter plot should exhibit key structural elements of a dataset,
such as clusters, hierarchy, sparsity, and outliers. 2-D projections are often used in fraud or anomaly
detection to find outlying entities, like people, transactions, or computers, or unusual clusters of
entities.

Model specific or model agnostic: Model agnostic; visualizing complex datasets with many

**Trust and understanding**: Projections add a degree of trust if they are used to confirm machine
learning modeling results. For instance, if known hierarchies, classes, or clusters exist in training or
test datasets and these structures are visible in 2-D projections, it is possible to confirm that a machine
learning model is labeling these structures correctly. A secondary check is to confirm that similar
attributes of structures are projected relatively near one another and different attributes of structures
are projected relatively far from one another. Consider a model used to classify or cluster marketing
segments. It is reasonable to expect a machine learning model to label older, richer customers
differently than younger, less affluent customers, and moreover to expect that these different groups
should be relatively disjointed and compact in a projection, and relatively far from one another.

# Import Required Libraries
Import the necessary libraries, including sklearn, matplotlib, and seaborn.

```python
# Importing necessary libraries
from sklearn import datasets
from sklearn.decomposition import PCA
from sklearn.manifold import MDS
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

# Load the Iris Dataset
Load the Iris dataset from sklearn.datasets.

```python
# Load the Iris dataset
iris = datasets.load_iris()

# The iris dataset is now loaded and can be used in subsequent sections.
```

# Data Preprocessing
Preprocess the data if necessary, such as normalizing the features.

```python
# Importing necessary library for data preprocessing
from sklearn.preprocessing import StandardScaler

# Separating out the features
x = iris.data
# Separating out the target
y = iris.target
# Standardizing the features
x = StandardScaler().fit_transform(x)
```

# Principal Components Analysis (PCA)
Apply PCA to the dataset and visualize the results.

```python
# Apply PCA to the dataset
pca = PCA(n_components=2)
principalComponents = pca.fit_transform(x)

# Create a DataFrame for the principal components
df_principal = pd.DataFrame(data = principalComponents, columns = ['principal component 1', 'principal component 2'])

# Add the target variable to the DataFrame
df_principal['target'] = y

# Visualize the results
plt.figure(figsize=(8, 6))
sns.scatterplot(x='principal component 1', y='principal component 2', hue='target', data=df_principal, palette=sns.color_palette("hsv", 3))

plt.title('2 component PCA')
plt.show()
```

# Multidimensional Scaling (MDS)
Apply MDS to the dataset and visualize the results.

```python
# Apply MDS to the dataset
mds = MDS(n_components=2)
mdsComponents = mds.fit_transform(x)

# Create a DataFrame for the MDS components
df_mds = pd.DataFrame(data = mdsComponents, columns = ['MDS component 1', 'MDS component 2'])

# Add the target variable to the DataFrame
df_mds['target'] = y

# Visualize the results
plt.figure(figsize=(8, 6))
sns.scatterplot(x='MDS component 1', y='MDS component 2', hue='target', data=df_mds, palette=sns.color_palette("hsv", 3))

plt.title('2 component MDS')
plt.show()
```

# t-Distributed Stochastic Neighbor Embedding (t-SNE)
Apply t-SNE to the dataset and visualize the results.

```python
import warnings
warnings.filterwarnings('ignore', module='threadpoolctl')

# Apply t-SNE to the dataset
tsne = TSNE(n_components=2)
tsneComponents = tsne.fit_transform(x)

# Create a DataFrame for the t-SNE components
df_tsne = pd.DataFrame(data = tsneComponents, columns = ['t-SNE component 1', 't-SNE component 2'])

# Add the target variable to the DataFrame
df_tsne['target'] = y

# Visualize the results
plt.figure(figsize=(8, 6))
sns.scatterplot(x='t-SNE component 1', y='t-SNE component 2', hue='target', data=df_tsne, palette=sns.color_palette("hsv", 3))

plt.title('2 component t-SNE')
plt.show()
```
